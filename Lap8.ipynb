```python
def strcat(s1,s2):
    return s1 +s2
```

```python
def fibo_string(n):
    fibo = strcat(['abc', 'def'] , [''] * (n - 1))
    for i in range(2, n + 1):
        fibo[i] = strcat(fibo[i - 1], fibo[i - 2])
    return fibo[n]
```

```python
test = fibo_string(2)
print(test)
```

```python
def fibo_char_at_k(n, k):
    fibo_str = fibo_string(n)
    if k < len(fibo_str):
        return fibo_str[k-1]
    else:
        return "Index out of range"
```

```python
k=3
n=2
print(f'The {k}-th character of the f{n}: {fibo_char_at_k(n,k)}')
```

```python
# Cập nhật lại hàm để tính đếm số lượng phép so sánh và gán
def fibo_string(n):
    global compare_count, assign_count
    fibo = strcat(['abc', 'def'] , [''] * (n - 1))
    assign_count += n + 1 
    for i in range(2, n + 1):
        fibo[i] = strcat(fibo[i - 1], fibo[i - 2])
        compare_count +=1
        assign_count += 1
    compare_count +=1 #break  
    return fibo[n]
```

```python
import matplotlib.pyplot as plt
# Khởi tạo danh sách để lưu lại số phép gán và so sánh cho mỗi giá trị của n
assign_counts_list = []
compare_counts_list = []
n_values = range(1, 11)
for n in n_values:
    assign_count = 0
    compare_count = 0
    fibo_string(n)
    assign_counts_list.append(assign_count)
    compare_counts_list.append(compare_count)

# Vẽ đồ thị
plt.plot(n_values, assign_counts_list, label='Assignments')
plt.plot(n_values, compare_counts_list, label='Comparisons')
plt.xlabel('n')
plt.ylabel('Counts')
plt.title('Number of Assignments and Comparisons in fibo_string')
plt.legend()
plt.show()
```

=> Độ phức tạp của thuật toán sử dụng là O(n)
